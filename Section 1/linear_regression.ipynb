```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
data
```

```python
data.describe()
```

```python
X = pd.DataFrame(data['production_budget_usd'])
y = pd.DataFrame(data['worldwide_gross_usd'])
```

```python
plt.scatter(X, y, alpha=0.3)
```

```python
regression = LinearRegression().fit(X, y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red')
plt.ylim(0, y.max()[0]*1.02)
plt.xlim(0, X.max()[0]*1.02)
plt.ylabel('Wolrdwide gross revenue ($)')
plt.xlabel('Production budget ($)')
plt.legend(['prediction', 'data'])
```

```python
plt.title('Zooming in prediction intercept with x and y axes')
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red')
plt.ylim(-8e6, 1e6)
plt.xlim(0, 3e6)
plt.ylabel('Wolrdwide gross revenue ($)')
plt.xlabel('Production budget ($)')
plt.legend(['prediction', 'data'])
```
